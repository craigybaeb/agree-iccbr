# Monotonicity (Real Data)

Effect of final score as each term increases using real data.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import plotly.express as px

ROOT = Path('..').resolve()
path = ROOT / 'results' / 'monotonicity.csv'
df = pd.read_csv(path)
df.head()
```

```python
import numpy as np
from scipy import stats
from sklearn.isotonic import IsotonicRegression

def scatter_with_trend(x_col, label):
    fig = px.scatter(df, x=x_col, y='R_bounded', trendline='ols', opacity=0.3,
                     title=f'R_bounded vs {label} (scatter + OLS)')
    fig.show()

def rank_correlation(x_col, label):
    rho, p = stats.spearmanr(df[x_col], df['R_bounded'])
    print(f'{label}: Spearman rho={rho:.4f}, p={p:.4g}')

def isotonic_fit(x_col, label):
    x = df[x_col].to_numpy()
    y = df['R_bounded'].to_numpy()
    order = np.argsort(x)
    x_sorted = x[order]
    y_sorted = y[order]
    iso = IsotonicRegression(increasing=True, out_of_bounds='clip')
    y_iso = iso.fit_transform(x_sorted, y_sorted)
    fig = px.line(x=x_sorted, y=y_iso, title=f'Isotonic fit: R_bounded vs {label}')
    fig.show()

for col, label in [
    ('S_plus', 'S_plus'),
    ('S_minus_u', 'S_minus_u'),
    ('one_minus_S_minus_x', '1 - S_minus_x'),
]:
    scatter_with_trend(col, label)
    rank_correlation(col, label)
    isotonic_fit(col, label)
```
