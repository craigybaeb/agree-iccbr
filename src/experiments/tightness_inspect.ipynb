# Tightness Results Inspection

This notebook loads CSV outputs from `experiments/tightness.py` and plots distribution comparisons.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

ROOT = Path('..').resolve()
RESULTS_DIR = ROOT / 'results'
print('RESULTS_DIR:', RESULTS_DIR)
```

```python
# Pick a tightness CSV to inspect
files = sorted(RESULTS_DIR.glob('tightness_*_*.csv'))
files
```

```python
path = files[0]
df = pd.read_csv(path)
df.head()
```

```python
fig = px.violin(
    df,
    x='scenario',
    y='R_bounded',
    color='scenario',
    box=True,
    points='all',
    title=f'Tightness (R_bounded) Distributions: {path.name}',
)
fig.show()
```

```python
summary_path = Path(str(path).replace('.csv', '_summary.csv'))
summary_df = pd.read_csv(summary_path)
summary_df
```
