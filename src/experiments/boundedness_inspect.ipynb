# Boundedness Results Inspection

This notebook loads CSV outputs from `experiments/boundedness.py` and plots violation rates.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

ROOT = Path('..').resolve()
RESULTS_DIR = ROOT / 'results'
print('RESULTS_DIR:', RESULTS_DIR)
```

```python
# Pick a summary CSV to inspect
summary_files = sorted(RESULTS_DIR.glob('boundedness_*_summary.csv'))
summary_files
```

```python
summary_path = summary_files[0]
summary_df = pd.read_csv(summary_path)
summary_df
```

```python
long_df = summary_df.melt(
    id_vars=['metric', 'n'],
    value_vars=['pct_any', 'pct_S_plus', 'pct_S_minus', 'pct_R_bounded'],
    var_name='violation',
    value_name='pct',
)
fig = px.bar(
    long_df,
    x='metric',
    y='pct',
    color='violation',
    barmode='group',
    title=f'Violation Rates: {summary_path.name}',
    range_y=[0, 1],
)
fig.show()
```
